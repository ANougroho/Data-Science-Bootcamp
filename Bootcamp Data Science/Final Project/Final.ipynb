```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Data_Negara_HELP.csv")
df
```

```python
df.describe()
```

```python
df_gdp = df.sort_values('GDPperkapita', ascending=True)
df_gdp.head()
```

```python
sns.barplot(data=df_gdp, x="GDPperkapita", y="Negara")
```

```python
df_inflasi = df.sort_values('Inflasi', ascending=False)
df_inflasi.head()
```

```python
sns.barplot(data=df_inflasi, x="Inflasi", y="Negara")
```

```python
df_harapan = df.sort_values('Harapan_hidup', ascending=True)
df_harapan.head()
```

```python
sns.barplot(data=df_harapan, x="Harapan_hidup", y="Negara")
```

```python
df_kematian = df.sort_values('Kematian_anak', ascending=False)
df_kematian.head()
```

```python
df_pendapatan = df.sort_values('Pendapatan', ascending=True)
df_pendapatan.head()
```

```python
sns.barplot(data=df_pendapatan, x="Pendapatan", y="Negara")
```

```python
df_fertiliti = df.sort_values('Jumlah_fertiliti', ascending=True)
df_fertiliti.head()
```

```python
sns.barplot(data=df_fertiliti, x="Jumlah_fertiliti", y="Negara")
```

```python
termiskin = pd.merge(pd.merge(df_gdp, df_pendapatan, on='Negara'),df_inflasi,on='Negara')
termiskin.head()
```

```python
sns.barplot(data=termiskin, x="GDPperkapita", y="Negara")
```

```python
kesehatan_min = pd.merge(pd.merge(df_harapan, df_kematian, on='Negara'),df_fertiliti,on='Negara')
kesehatan_min.head()
```

```python
sns.barplot(data=kesehatan_min, x="Harapan_hidup", y="Negara")
```

```python
negara_membutuhkan = pd.merge(termiskin, kesehatan_min, on='Negara')
negara_membutuhkan
```

```python
sns.barplot(data=negara_membutuhkan,x='GDPperkapita_y', y="Negara")
```
